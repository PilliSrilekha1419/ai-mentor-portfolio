Big-O notation is simply a way to describe how fast an algorithm grows when the input size increases.

In interviews, companies like Tata Consultancy Services, Infosys, and Cognizant usually don’t expect you to speak like a mathematician. They want to know whether you can judge:

Which solution is faster
Which solution scales better
Why one approach is better than another

Let’s understand it like we would in a placement prep session.

Imagine This

Suppose I give you an array of 10 numbers and ask you to find whether 25 exists.

You may check:

first element
second element
third element…

until you find 25.

If there are n elements, in the worst case you may check all n elements.

That is called:

O(n)

This means the time grows linearly with input size.

If input doubles, work roughly doubles.

Common Big-O Values You Must Know
1. Constant Time — Fastest

Example:
Accessing an array element using index.

arr[5]

No matter whether array size is 10 or 10 lakh, it takes almost same time.

O(1)

Interview line:

“Execution time does not depend on input size.”

2. Linear Time

Example:
Linear Search

for(int i=0; i<n; i++)

You may visit every element once.

O(n)

If n = 100, about 100 operations.
If n = 1000, about 1000 operations.

3. Logarithmic Time

Example:
Binary Search

Instead of checking every element, we divide the search space into half each time.

mid = (low + high)/2

O(logn)

Very important for interviews.

Example:

1000 elements → around 10 checks
1,000,000 elements → around 20 checks

That’s why binary search is powerful.

4. Quadratic Time

Nested loops.

for(i=0; i<n; i++)
{
   for(j=0; j<n; j++)
}

Each loop runs n times.

Total operations:

O(n
2
)

If n = 1000, operations become 10 lakh.

This becomes slow for large inputs.

How Interviewers Ask Big-O

They usually ask in practical ways.

Example 1

“What is the time complexity of linear search?”

Answer:

“Worst-case time complexity is O(n) because we may need to traverse the complete array.”

Example 2

“Why is binary search faster?”

Answer:

“Because it eliminates half of the search space in every iteration, giving O(log n) complexity.”

Example 3

“Which is better: O(n) or O(log n)?”

Answer:

“O(log n) is better for large datasets because growth is much slower.”

Easy Trick to Calculate Big-O
Single Loop
for(i=0; i<n; i++)

→ O(n)

Nested Loops
for(i=0; i<n; i++)
   for(j=0; j<n; j++)

→ O(n²)

Loop Dividing by 2
while(n>1)
{
   n = n/2;
}

→ O(log n)

Important Interview Concept

Big-O usually talks about the worst-case scenario.

Example:
In linear search:

Best case → element found first → O(1)
Worst case → element found last → O(n)

Companies mostly ask for worst-case complexity unless specified.

Real Placement Tip

For service-based companies, you usually won’t get very deep mathematical questions on Big-O.

But you should confidently know:

O(1)
O(log n)
O(n)
O(n²)

and identify them from loops.

That alone covers most rounds for 7–10 LPA fresher interviews.

For product companies, they may ask:

optimize this code,
reduce complexity,
compare approaches.
Final Shortcut You Should Remember
Complexity	Meaning	Performance
O(1)	Constant	Excellent
O(log n)	Divide & conquer	Very fast
O(n)	Linear	Acceptable
O(n²)	Nested iteration	Slow for large input
